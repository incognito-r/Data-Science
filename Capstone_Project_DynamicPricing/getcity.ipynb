```python
import pandas as pd
df = pd.read_csv('data/uber.csv')
df.head()
```

```python
df2 = df[['pickup_latitude', 'pickup_longitude', 'dropoff_latitude', 'dropoff_longitude']].sample(n=10)
```

```python
df2
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
from tqdm import tqdm

# Initialize geocoder
geolocator = Nominatim(user_agent="geoapi", timeout=10)

# Function to get city from latitude and longitude
def get_city(lat, lon):
    try:
        location = geolocator.reverse((lat, lon), exactly_one=True)
        if location and "address" in location.raw:
            return location.raw["address"].get("city", location.raw["address"].get("town", location.raw["address"].get("village", "Unknown")))
    except GeocoderTimedOut:
        return "Timeout"
    return "Unknown"

# Apply function with progress bar
tqdm.pandas()
df2["pickup_city"] = df2.progress_apply(lambda row: get_city(row["pickup_latitude"], row["pickup_longitude"]), axis=1)
df2["dropoff_city"] = df2.progress_apply(lambda row: get_city(row["dropoff_latitude"], row["dropoff_longitude"]), axis=1)

# Display the result
df2
```
