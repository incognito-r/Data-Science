```python
df['Month-Year'] = pd.to_datetime(df['Month-Year'], format="%b-%y")
```

```python
df.dtypes
```

```python
df['Month-Name'] = df['Month-Year'].dt.month_name()
df['Year'] = df['Month-Year'].dt.year
```

```python
df.head()
```

```python
df['Month-Name'].value_counts().index
```

```python
dt1=df.pivot_table(index="Year", columns="Month-Name", values="Number of Tractor Sold", aggfunc="sum")
dt2=dt1[['January', 'February', 'March', 'April', 'May', 'June', 'July',
       'August', 'September', 'October', 'November', 'December']]
dt2
dt2.plot(kind="line", figsize=(10,6))
```
