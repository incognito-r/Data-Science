```python
def outlier(x):
  x=x.clip(lower=x.quantile(0.01))
  x=x.clip(upper=x.quantile(0.99))
  return x
```

```python
df['income'] = outlier(df['income'])
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
std=StandardScaler().fit(df)
X_std=pd.DataFrame(std.transform(df), columns=df.columns)
# X_std
```

```python
km=KMeans(n_clusters=3,random_state=100)
km.fit(df)
```

```python
km.inertia_
```

```python
km.cluster_centers_[0]
```

```python
X_std.columns
```

```python
df_copy=df.copy()
```

```python
df_copy["cluster"]=km.labels_  # adding clusters to origional data set
```

```python
df_copy.head()
```

```python
df_copy["cluster"].value_counts()
```

```python
clust_summary=df_copy.groupby("cluster").mean().T.reset_index()
clust_summary.columns=["Features", "cluster0", "clutser1", "cluster2"]
clust_summary
```

```python
overall=df_copy.describe().T["mean"].reset_index()
overall.columns=["Features", "overall"]
overall
```

```python
final_summary=pd.merge(overall, clust_summary, on="Features", how="inner")
final_summary.to_excel("cluster_summary0.xlsx")
```
